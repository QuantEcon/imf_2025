# Temporal difference learning exercise

We consider a different environment with discrete actions/states.

The agent is a "store manager" who must choose how many units of a good to keep in stock.

* Demand is random and you earn \\$10 per good sold.
* The manager can order up to 5 units of the good each day that are delivered the following day.
  * Each order costs a \\$2 fixed cost plus \\$4 cost per unit ordered.
  * If there is demand for $N$ goods but you only have $M < N$ goods then you can only sell $M$ -- There is no instantaneous delivery.
* Any of the goods left over night incur a storage cost of \\$1.
* You can store a maximum of 15 units of the good.

**Question 1**: Write a Bellman equation describing the store manager's problem.

**Question 2**: If you solved this with both Sarsa and Q-learning, how would you expect the solutions to differ? Why?

**Question 3**: Write an environment that can simulate the inventory management environment. This environment should have the following methods:

* `reset`
* `step`

**Question 4**: Define a Sarsa agent and play the game for 30 days. How much profit does the manager earn?

**Question 5**: Allow the Sarsa manager to play for 1 additional year (on top of the 30 days) and learn. Now let the manager play for 30 more days. How much does the manager earn during this 30 days?

**Question 6**: Define a Q-learning agent and play the game for 30 days. How much profit does the manager earn?

**Question 7**: Allow the Q-learning manager to play for 1 additional year (on top of the 30 days) and learn. Now let the manager play for 30 more days. How much does the manager earn during this 30 days?

**Question 8**: How do the profits achieved in Question 5 and Question 7 differ? Does this appear to be randomness or is there an underlying reason?
